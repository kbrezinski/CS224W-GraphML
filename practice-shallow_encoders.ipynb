```python
import networkx as nx
import matplotlib.pyplot as plt

#from torch_geometric.datasets import KarateClub
```

```python
G = nx.karate_club_graph()
```

```python
weights = nx.get_edge_attributes(G,'weight').values()  # weights are missing for some reason
colors = ['r' if label == 'Mr. Hi' else 'g' for label in nx.get_node_attributes(G, 'club').values()]

nx.draw(G, pos=nx.kamada_kawai_layout(G), with_labels = True, width=list(weights), node_color=colors)
plt.show()
```

```python
# fetch the adjacency matrix
nx.adjacency_matrix(G)#.todense()
```

```python
# betweeness centrality - how many shortest paths pass through u?
sorted([(v, i) for i, v in nx.betweenness_centrality(G, normalized=False).items()], reverse=True)[:5]
```

```python
# eigenvector centrality
sorted([(v, i) for i, v in nx.eigenvector_centrality(G).items()], reverse=True)[:5]
```

```python
# katz centrality - centrality based on the centrality of its neigbors
sorted([(v, i) for i, v in nx.katz_centrality(G).items()], reverse=True)[:5]
```

```python
# closeness centrality - how central is a node? low paths between neigboring nodes
sorted([(v, i) for i, v in nx.closeness_centrality(G).items()], reverse=False)[:5]
```

```python
# clustering - how well connected v's neighbors are; smallest means it's a terminal node
sorted([(v, i) for i, v in nx.clustering(G).items()], reverse=True)[:5]
```

```python
# Jaccard Coefficient - which nodes have high degree of overlapping neighbors
sorted([(p, u, v) for u, v, p in nx.jaccard_coefficient(G)], reverse=True)[:5]
```

```python
# Adamic Adar - 
sorted([(p, u, v) for u, v, p in nx.adamic_adar_index(G)], reverse=True)[:5]
```
