```python
import networkx as nx
import matplotlib.pyplot as plt

#from torch_geometric.datasets import KarateClub
```

```python
G = nx.karate_club_graph()
```

```python
weights = nx.get_edge_attributes(G,'weight').values()
colors = ['r' if label == 'Mr. Hi' else 'g' for label in nx.get_node_attributes(G, 'club').values()]

nx.draw(G, pos=nx.kamada_kawai_layout(G), with_labels = True, width=list(weights), node_color=colors)
plt.show()
```
